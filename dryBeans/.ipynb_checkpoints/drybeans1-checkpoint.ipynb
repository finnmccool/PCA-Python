```python
def write_versions(versions):
    from pathlib import Path
    if Path('pythonVersions.txt').is_file():
        print ("File exists with these versions.",end='\n\n')
    else:
        with open ('pythonVersions.txt', 'w') as f:  
            for key, value in versions.items():  
                f.write('%s:%s\n' % (key, value))
```

Use <strong>set_axis_off</strong> to get rid of the X & Y axis points.

```python
def show_img(img, bean,pos):
    '''img = numpy array from npimg.read
       bean = string for the plot title
       pos = integer column position for plot
       returns nothing plots the image'''
    ax = fig.add_subplot(1, 3, pos)
    imgplot = plt.imshow(img)
    ax.set_title(bean)
    ax.set_axis_off()
    fig.add_axes(ax)
```

<h2>Principal Component Analysis</h2>

<ul>PCA is a technique used for the identification of a smaller number of uncorrelated variables known as principal components from a larger set of data. The beans data set has 12 variables to identify the target bean. PCA will help us to identify which variables have most effect on classifying our bean.<br><br>"Instead of your original variables, you get the same number of new composite variables back. The difference being that before you performed the PCA, the variables you had might be correlated to one another. What PCA attempts is to produce completely uncorrelated variables which are called Principal Components. Furthermore, the Principal Components themselves are ordered such that the first component captures the largest amount of variance in the data set as possible, the second the next highest level of variance and so on."<br><br>This is a quote from the article below which is well worth a read.</ul>

```python
DataSource = IFrame(src='https://firefinch.io/a-simple-introduction-to-pca/', width=1000, height=200)
display(DataSource)
```

We will use the **df_minus_shape_factor** for the values and **df.bean** for the ***target*** values.<br>
The data need to be scaled first. This is important. Variables with big values will have more influence. They all need to be singing from the same hymn book as it were. Use the **sklearn** library and its modules. I will be using StandardScaler and PCA modules. there are various names for scaling which are all slightly different methods. Look out for words like *normalizing*, *standardization*. Look them up and find out what the differences are. ***StandardScaler*** as used here is akin to a Z score. The formula being x<sub>standardised</sub> = (x - mean) / std<br> See below for a visual description

![alt text](../img/zscore.png)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

One of the great things about modern IDE's is that you can do away with variable names like **X,y,x,Y** which I find difficult to distinguish on the screen. If you can give a variable a recognisable name then use it. Your IDE will prompt you. You will see many tutorials usin X or y. Copy and use them if you want, but I prefer real names! I will use some inline comments now to help guide me through the PCA process.

<h3>Scaling the data</h3>

```python
##First - create a numpy array of the feature varaibles
##In this case we have columns 1-12. NB the index column is removed as well as the "bean" classification
bean_features = df_minus_shapefactor[df_minus_shapefactor.columns[1:12]].values

##The target column is the "bean" column
target = df_minus_shapefactor[df_minus_shapefactor.columns[12:13]].values

##Now we standardise the values into the numpy array - bean_features_standard
bean_features_standard = StandardScaler().fit_transform(bean_features)
```

It is worthwhile to show 2 plots of how we have manipulated the data to help understand what we have just done.

```python
fig,axs = plt.subplots(1,2,figsize=(15, 4), sharex=False, sharey=False)
fig.suptitle('Standardized V Original KDE',fontsize=20, y=1)
sns.kdeplot(ax=axs[0], data=bean_features_standard).set_title('Standardized', fontdict=fontdict1)
sns.kdeplot(ax=axs[1], data=bean_features).set_title('Original', fontdict=fontdict1);
```

This shows how the data has been centered and scaled!!

You will see examples using **X** and **y** for making analyses where ever you see examples on the internet. In our case **bean_features_standard** will be **X** and **target** will be **y**.<br>I hope using names for the variables will make what we are doing a bit more clear. Remember. **bean_features_standard** is just the values of each figure from **df_minus_shapefactor** standardises using StandardScaler. **target** is just the bean types olumn from the same DF.

<h3>Calculating PCA</h3>

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=11)

principalComponents = pca.fit_transform(bean_features_standard)

df_PCA12 = pd.DataFrame(data = principalComponents
             , columns = ['pca_1', 'pca_2','pca_3','pca_4','pca_5','pca_6','pca_7','pca_8','pca_9' ,'pca_10','pca_11'])
```

```python
df_PCA12['bean'] = df_minus_shapefactor['bean']
```

```python
df_PCA12.describe()
```

<h3>Explained Variance</h3>

There are 2 methods of determining the explained variance. One from the pca library and one from the pandas library. How do they compare?

```python
explained_variance = pca.explained_variance_
explained_variance.round(2)
```

```python
df_PCA12.var().plot(kind ='bar',alpha=0.7)
plt.plot(explained_variance, color='g',marker='o');
```

```python
pca_variance = df_PCA12.var()
for i in range(0,11):
    print(f'PCA{i+1} accounts for {round(pca_variance[i]*10,1)}% of variance')
```

<h3>Outliers</h3>

You may notice that the maximum value for PCA 3 is 13.6. This looks like a high outlier. We can have a look at the distributions for PCA 3.

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.grid()
sns.kdeplot(data=df_PCA12, x='pca_3', hue='bean');
```

PCA 3 has a few outliers. Lets see how many are above 4

```python
len(df_PCA12.query('pca_3 > 4.0'))
```

So 77 data points out of 13k + are above 4.0. I think we can get rid of these as they are definitely outliers.

```python
df_PCA12 = df_PCA12[df_PCA12['pca_3'] <= 4.0]
```

<h3>3D Plotting</h3>

```python
from mpl_toolkits.mplot3d import Axes3D
```

Note the magic vommand ***%matplotlib*** you can add ***qt*** as well. All that does, is show the plot in a new window where you can manipulate it in 3D.<br> remember to use ***%matplotlib inline*** to switch it back!!

```python
%matplotlib

fig = plt.figure(figsize=(12,12))
ax = Axes3D(fig)
x = df_PCA12['pca_1']
y = df_PCA12['pca_2']
z = df_PCA12['pca_3']
ax.set_title('Bean Data set PCA', size=14)
ax.scatter3D(x,y,z,s=0.5)
ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')
ax.set_zlabel('PCA 3')
ax.w_xaxis.set_ticklabels(())
ax.w_yaxis.set_ticklabels(())
ax.w_zaxis.set_ticklabels(())
plt.show();
```

```python
%matplotlib

fig = plt.figure(figsize = (12,12))
ax = Axes3D(fig)   
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)

cmap1 = plt.get_cmap('PiYG_r')

sctt = ax.scatter3D(x, y, z,
                    alpha = 0.8,
                    c = (x + y + z),
                    cmap = cmap1,
                    s=0.8)
 
plt.title('3D Plot of PCA 1 2 and 3')
ax.set_xlabel('PCA 1', fontweight ='bold')
ax.set_ylabel('PCA 2', fontweight ='bold')
ax.set_zlabel('PCA 3', fontweight ='bold')

plt.show()
```

Lets have a look at PCA 1 against PCA2 and see if there is good ***'bean'*** separation.

```python
%matplotlib inline
plt.figure(figsize=(10,10))
plt.grid()
plt.title('PCA 1 V PCA 2 Using Bean for Color Separation', fontsize=20)
sns.scatterplot(data=df_PCA12, y='pca_1', x='pca_2', hue='bean')
plt.show()
```

<h3>PCA reduces dimensionality</h3>

Using PCA has enabled the reduction of 11 variables down to the 2 or 3 which describe the most variation between our different beans.<br>Imagine you have a data set with 30 or 100 variable. <h2>Multivariate Regression from PCA Analysis</h2>

First we have to make a copy of outr PCA DF and remove everything but PCA 1,2 and 3 and keep our categorical variable, ***'bean'***.<br>This can be done in one line by dropping the columns we don't need in one DF and assigning the result to  new DF.<br>The trick is to use **iloc** and select the columns by **[:,3:12]**.<br>Remember python is zero counting and the last column (12) actually stops at 11. Just to confuse the unwary!!

```python
df_PCA123 = df_PCA12.drop(df_PCA12.iloc[:,3:11], axis=1)
df_PCA123.head()
```

<h3>Label Encoding</h3>

Now we have to decide what to do with the categorical variable. Machine Learning does not deal with these. It expects numbers.<br>Lets look at label encoding first.<br>Bean labels need a number code!

```python
#How many observations per bean have we got??
df_PCA123.bean.value_counts()
```

Use the ***LabelEncoder*** module from ***sklearn*** and give each bean a numerical code.

```python
from sklearn.preprocessing import LabelEncoder
bean_label_encoder = LabelEncoder()
df_PCA123['bean_id'] = bean_label_encoder.fit_transform(df_PCA123['bean'])
df_PCA123.bean_id.value_counts()
```

You can use ***sample*** in the same way as ***head*** or ***tail*** to take random samples from the ***DF*** to check the success of the label encoding.

```python
df_PCA123.sample(5)
```
